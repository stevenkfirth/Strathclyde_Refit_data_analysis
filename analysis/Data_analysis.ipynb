# Data analysis of the Strathclyde REFIT data

A notebook which demonstrates analysis of the Strathclyde dataset

## Setup

```python
import pandas as pd
import os
fp=r'C:\Users\cvskf\OneDrive - Loughborough University\_Data\REFIT_Electrical_load_measurements_(cleaned)\Data\CLEAN_REFIT_081116'
```

## Data processing

This loads the data, resamples it to half hour intervals and saves the new data files.

```python
flag=True
for filename in os.listdir(fp):
    print(filename)
    path=os.path.join(fp,filename)
    df=pd.read_csv(path,parse_dates=['Time'])
    df=df.set_index('Time')
    df=df.resample('30T').mean()
    df=df.reset_index()
    df['file']=filename
    if flag:
        df2=df
        flag=False
    else:
        df2=df2.append(df)
df2.to_csv('NewData.csv')
```
